```python
# Import qilish
from ultralytics import YOLO
from pathlib import Path
import os
import gc

print("Ultralytics muvaffaqiyatli import qilindi.")
```

```python
# Modellar saqlanadigan papkani yaratish
MODELS_FOLDER = Path("downloaded_models").resolve()
MODELS_FOLDER.mkdir(parents=True, exist_ok=True)

print("Modellar quyidagi papkaga saqlanadi:")
print(MODELS_FOLDER)
```

```python
# Object detection modellari ro‘yxati

MODEL_NAMES = [
    # YOLOv5u
    "yolov5nu.pt",
    "yolov5su.pt",
    "yolov5mu.pt",
    "yolov5lu.pt",
    "yolov5xu.pt",

    # YOLOv8
    "yolov8n.pt",
    "yolov8s.pt",
    "yolov8m.pt",
    "yolov8l.pt",
    "yolov8x.pt",

    # YOLO11
    "yolo11n.pt",
    "yolo11s.pt",
    "yolo11m.pt",
    "yolo11l.pt",
    "yolo11x.pt",

    # YOLO26
    "yolo26n.pt",
    "yolo26s.pt",
    "yolo26m.pt",
    "yolo26l.pt",
    "yolo26x.pt",
]

print("Modellar soni:", len(MODEL_NAMES))
```

```python
#Barcha modellarni yuklash

old_directory = Path.cwd()

downloaded_models = []
existing_models = []
failed_models = []

try:
    # Modellar downloaded_models papkasiga tushishi uchun
    os.chdir(MODELS_FOLDER)

    for number, model_name in enumerate(MODEL_NAMES, start=1):
        model_path = Path(model_name)

        print("\n" + "=" * 70)
        print(f"{number}/{len(MODEL_NAMES)} — {model_name}")
        print("=" * 70)

        if model_path.exists():
            size_mb = model_path.stat().st_size / (1024 * 1024)

            print(f"Oldindan mavjud: {model_name}")
            print(f"Hajmi: {size_mb:.2f} MB")

            existing_models.append(model_name)
            continue

        model = None

        try:
            print(f"Yuklanmoqda: {model_name}")

            # Model mavjud bo‘lmasa, internetdan avtomatik yuklanadi
            model = YOLO(model_name)

            if model_path.exists():
                size_mb = model_path.stat().st_size / (1024 * 1024)

                print(f"Muvaffaqiyatli yuklandi: {model_name}")
                print(f"Hajmi: {size_mb:.2f} MB")
                print(f"Task: {model.task}")

                downloaded_models.append(model_name)
            else:
                print("Yuklash bajarildi, lekin model fayli topilmadi.")
                failed_models.append(model_name)

        except Exception as error:
            print(f"Xato yuz berdi: {model_name}")
            print(error)

            failed_models.append(model_name)

        finally:
            del model
            gc.collect()

finally:
    # Notebookning dastlabki papkasiga qaytish
    os.chdir(old_directory)

print("\n" + "=" * 70)
print("YUKLASH YAKUNI")
print("=" * 70)

print("Yangi yuklangan:", len(downloaded_models))
print("Oldindan mavjud:", len(existing_models))
print("Yuklanmagan:", len(failed_models))

if failed_models:
    print("\nYuklanmagan modellar:")
    for model_name in failed_models:
        print("-", model_name)

# Bu cell barcha modellarni ketma-ket yuklaydi. Katta l va x modellar sababli yuklash uzoqroq davom etishi va
# ko‘p disk joyi talab qilishi mumkin.
```

```python
# Yuklangan fayllarni tekshirish

model_files = sorted(MODELS_FOLDER.glob("*.pt"))

print(f"Topilgan modellar soni: {len(model_files)}")
print("-" * 70)

for model_path in model_files:
    size_mb = model_path.stat().st_size / (1024 * 1024)

    print(f"{model_path.name:<20} {size_mb:>10.2f} MB")
```
