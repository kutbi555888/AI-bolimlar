```python
from pathlib import Path
import random
import yaml
import cv2
import matplotlib.pyplot as plt

DATASET_DIR = Path('new_lesson-1-yolov8').resolve()

# Fallback absolute path, useful if notebook starts from another folder.
if not DATASET_DIR.exists():
    DATASET_DIR = Path(r'/content/bike-2-zy9qw-1-yolov8').resolve()

DATA_YAML = DATASET_DIR / 'data.yaml'

print('Dataset folder:', DATASET_DIR)
print('data.yaml:', DATA_YAML)
print('Dataset exists:', DATASET_DIR.exists())
print('YAML exists:', DATA_YAML.exists())
```

```python
with open(DATA_YAML, 'r', encoding='utf-8') as f:
    data_cfg = yaml.safe_load(f)

print(DATA_YAML.read_text(encoding='utf-8'))

names = data_cfg['names']
if isinstance(names, dict):
    names = [names[k] for k in sorted(names)]

print('Class names:', names)
```

```python
train_cfg = dict(data_cfg)
train_cfg['path'] = str(DATASET_DIR)
train_cfg['train'] = 'train/images'
train_cfg['val'] = 'valid/images'
train_cfg['test'] = 'test/images'

TRAIN_YAML = DATASET_DIR / 'data_train_fixed.yaml'
with open(TRAIN_YAML, 'w', encoding='utf-8') as f:
    yaml.safe_dump(train_cfg, f, sort_keys=False, allow_unicode=True)

print('Training YAML:', TRAIN_YAML)
print(TRAIN_YAML.read_text(encoding='utf-8'))
```

```python
sample_label = next((DATASET_DIR / 'train' / 'labels').glob('*.txt'))
print('Label file:', sample_label.name)
print(sample_label.read_text(encoding='utf-8'))
```

```python
IMG_EXTS = {'.jpg', '.jpeg', '.png', '.bmp', '.webp'}

def get_images(split='train'):
    if split == 'val':
        split = 'valid'
    image_dir = DATASET_DIR / split / 'images'
    return sorted([p for p in image_dir.rglob('*') if p.suffix.lower() in IMG_EXTS])

def label_path_for_image(image_path):
    image_path = Path(image_path)
    return image_path.parent.parent / 'labels' / f'{image_path.stem}.txt'

def yolo_to_xyxy(box, width, height):
    x, y, bw, bh = box
    x1 = int((x - bw / 2) * width)
    y1 = int((y - bh / 2) * height)
    x2 = int((x + bw / 2) * width)
    y2 = int((y + bh / 2) * height)
    x1 = max(0, min(x1, width - 1))
    y1 = max(0, min(y1, height - 1))
    x2 = max(0, min(x2, width - 1))
    y2 = max(0, min(y2, height - 1))
    return x1, y1, x2, y2

def draw_labels(image_path):
    img_bgr = cv2.imread(str(image_path))
    if img_bgr is None:
        raise ValueError(f'Image not found: {image_path}')

    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    h, w = img.shape[:2]
    label_path = label_path_for_image(image_path)
    lines = label_path.read_text(encoding='utf-8').splitlines() if label_path.exists() else []

    for line in lines:
        parts = line.split()
        if len(parts) < 5:
            continue
        cls_id = int(float(parts[0]))
        box = list(map(float, parts[1:5]))
        x1, y1, x2, y2 = yolo_to_xyxy(box, w, h)
        label = names[cls_id] if 0 <= cls_id < len(names) else str(cls_id)

        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 70, 30), 3)
        cv2.putText(img, label, (x1, max(25, y1 - 8)), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 70, 30), 2)

    return img, len(lines)

def show_labeled_samples(split='train', n=6):
    images = get_images(split)
    if not images:
        raise FileNotFoundError(f'No images found for split: {split}')

    chosen = random.sample(images, min(n, len(images)))
    cols = 3
    rows = (len(chosen) + cols - 1) // cols
    plt.figure(figsize=(15, 5 * rows))

    for i, image_path in enumerate(chosen, 1):
        img, box_count = draw_labels(image_path)
        plt.subplot(rows, cols, i)
        plt.imshow(img)
        plt.title(f'{image_path.name}\nboxes: {box_count}', fontsize=9)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

print('Train images:', len(get_images('train')))
print('Valid images:', len(get_images('valid')))
print('Test images:', len(get_images('test')))

show_labeled_samples('train', n=6)
```

```python
!pip install ultralytics
from ultralytics import YOLO
import torch

DEVICE = 0 if torch.cuda.is_available() else 'cpu'
SAMPLE_IMAGE = (get_images('test') or get_images('valid') or get_images('train'))[0]

base_model = YOLO('yolov8n.pt')
base_pred = base_model.predict(str(SAMPLE_IMAGE), conf=0.25, device=DEVICE, verbose=False)[0]

plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(base_pred.plot(), cv2.COLOR_BGR2RGB))
plt.title(f'Before fine-tuning | boxes: {len(base_pred.boxes)}')
plt.axis('off')
plt.show()

print('Test image:', SAMPLE_IMAGE)
```

# Yolo modelni fine tuning qilib o'zimizga moslashtirish

```python
model = YOLO('yolov8n.pt')

train_results = model.train(
    data=str(TRAIN_YAML),
    epochs=5,
    imgsz=640,
    device=DEVICE,
    project='runs/simple_lesson10',
    name='finetune_5_epochs',
    exist_ok=True
)

BEST_MODEL = Path(train_results.save_dir) / 'weights' / 'best.pt'
print('Best model:', BEST_MODEL)
```

```python
fine_model = YOLO(str(BEST_MODEL))
fine_pred = fine_model.predict(str(SAMPLE_IMAGE), conf=0.25, device=DEVICE, verbose=False)[0]

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].imshow(cv2.cvtColor(base_pred.plot(), cv2.COLOR_BGR2RGB))
axes[0].set_title(f'Before fine-tuning | boxes: {len(base_pred.boxes)}')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(fine_pred.plot(), cv2.COLOR_BGR2RGB))
axes[1].set_title(f'After fine-tuning | boxes: {len(fine_pred.boxes)}')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
metrics = fine_model.val(data=str(TRAIN_YAML), split='val', imgsz=640, device=DEVICE)

print(f'mAP50: {metrics.box.map50:.4f}')
print(f'mAP50-95: {metrics.box.map:.4f}')
print(f'Precision: {metrics.box.mp:.4f}')
print(f'Recall: {metrics.box.mr:.4f}')
```

```python
test_images = get_images('test')
if not test_images:
    test_images = get_images('valid')

test_images = test_images[:6]
test_preds = fine_model.predict([str(p) for p in test_images], conf=0.25, device=DEVICE, verbose=False)

cols = 3
rows = (len(test_preds) + cols - 1) // cols
plt.figure(figsize=(15, 5 * rows))

for i, pred in enumerate(test_preds, 1):
    plt.subplot(rows, cols, i)
    plt.imshow(cv2.cvtColor(pred.plot(), cv2.COLOR_BGR2RGB))
    plt.title(f'Test prediction | boxes: {len(pred.boxes)}', fontsize=9)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# data papkani zip qilib  Google Drivega ham, kompyuterga ham tayyorlash

# Quyidagi kod:

# papkani topadi;
# ZIP yaratadi;
# Drive’ga nusxalaydi;
# kompyuterga yuklab olishga tayyorlaydi.
from pathlib import Path
import shutil

from google.colab import drive


# ============================================================
# 1. Google Drive’ni ulash
# ============================================================

drive.mount(
    "/content/drive"
)


# ============================================================
# 2. Dataset papkasi
# ============================================================

SOURCE_FOLDER = Path(
    "/content/bike-2-zy9qw-1-yolov8"
)


if not SOURCE_FOLDER.exists():
    possible_folders = list(
        Path("/content").glob(
            "bike-2-zy9qw-1-yolov8*"
        )
    )

    if not possible_folders:
        raise FileNotFoundError(
            "bike-2-zy9qw-1-yolov8 "
            "papkasi topilmadi."
        )

    SOURCE_FOLDER = possible_folders[0]


print("Topilgan dataset:")
print(SOURCE_FOLDER)


# ============================================================
# 3. ZIP fayl yaratish
# ============================================================

LOCAL_ZIP_BASE = Path(
    "/content"
) / SOURCE_FOLDER.name


ZIP_FILE_PATH = shutil.make_archive(
    base_name=str(LOCAL_ZIP_BASE),
    format="zip",
    root_dir=str(SOURCE_FOLDER.parent),
    base_dir=SOURCE_FOLDER.name,
)


ZIP_FILE_PATH = Path(
    ZIP_FILE_PATH
)


# ============================================================
# 4. ZIP faylni Drive’ga nusxalash
# ============================================================

DRIVE_FOLDER = Path(
    "/content/drive/MyDrive/"
    "YOLO_Datasets"
)

DRIVE_FOLDER.mkdir(
    parents=True,
    exist_ok=True,
)


DRIVE_ZIP_PATH = (
    DRIVE_FOLDER
    / ZIP_FILE_PATH.name
)


shutil.copy2(
    ZIP_FILE_PATH,
    DRIVE_ZIP_PATH,
)


# ============================================================
# 5. Natija
# ============================================================

print("\n" + "=" * 70)
print("SAQLASH TUGADI")
print("=" * 70)

print("\nLokal ZIP:")
print(ZIP_FILE_PATH)

print("\nGoogle Drive’dagi ZIP:")
print(DRIVE_ZIP_PATH)

print(
    "\nZIP hajmi:",
    round(
        ZIP_FILE_PATH.stat().st_size
        / (1024 * 1024),
        2,
    ),
    "MB",
)

print(
    "\nDrive’ga saqlandimi:",
    DRIVE_ZIP_PATH.exists(),
)
```
