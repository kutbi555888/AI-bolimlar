# Image Classification turlari

Ushbu notebook image classification’ning 3 ta asosiy turini sodda usulda tushuntiradi:

1. **Binary classification** -> Ha / Yo‘q
2. **Multi-class classification** -> 1 ta class tanlash
3. **Multi-label classification** -> Ko‘p label tanlash

Shuningdek, har bir tur uchun yaxshi dataset misolini ham beradi.

## 1. Binary Classification

Binary classification degani model faqat **2 ta class** orasidan birini tanlaydi.

Misollar:

- Cat yoki Dog
- Normal yoki Diseased
- Ha yoki Yo‘q

### Oddiy tushuncha

Model bitta savolga javob beradi:

- Bu image class A ga tegishlimi?
- Yoki class B ga tegishlimi?

```python
# Binary classification example labels
binary_classes = ["cat", "dog"]

print("Binary classes:", binary_classes)
print("Number of classes:", len(binary_classes))
```

### Yaxshi Binary Dataset

**Cats vs Dogs**

- Type: Binary classification
- Classes: `cat`, `dog`
- Nima uchun yaxshi: beginnerlar uchun tushunish juda oson
- Rasmiy manba: Microsoft / TensorFlow Datasets

Links:

- Microsoft download page: https://www.microsoft.com/en-gb/download/details.aspx?id=54765
- TensorFlow Datasets page: https://www.tensorflow.org/datasets/catalog/cats_vs_dogs

## 2. Multi-class Classification

Multi-class classification degani model **ko‘p classlar orasidan bitta class** tanlaydi.

Muhim qoida:

- Image faqat **bitta to‘g‘ri class**ga tegishli bo‘ladi

Misollar:

- airplane
- car
- bird
- dog
- ship

### Oddiy tushuncha

Model **faqat bitta javob** tanlashi kerak.

```python
# Multi-class classification example labels
multiclass_classes = [
    "airplane",
    "automobile",
    "bird",
    "cat",
    "deer",
    "dog",
    "frog",
    "horse",
    "ship",
    "truck",
]

print("Multi-class classes:", multiclass_classes)
print("Number of classes:", len(multiclass_classes))
```

### Yaxshi Multi-class Dataset

**CIFAR-10**

- Type: Multi-class classification
- Number of classes: 10
- Rule: faqat **1** ta class tanlash
- Nima uchun yaxshi: kichik, mashhur va o‘rgatish uchun oson
- Rasmiy manba: University of Toronto / TensorFlow Datasets

Links:

- Official dataset page: https://www.cs.utoronto.ca/~kriz/cifar.html
- TensorFlow Datasets page: https://www.tensorflow.org/datasets/catalog/cifar10

## 3. Multi-label Classification

Multi-label classification degani model bir vaqtning o‘zida **ko‘p label** tanlashi mumkin.

Bu multi-class classification’dan farq qiladi.

Multi-class’da:

- faqat 1 ta label tanlanadi

Multi-label’da:

- kerak bo‘lsa ko‘p label tanlanadi

### Misol

Bitta image ichida quyidagilar bo‘lishi mumkin:

- person
- bicycle
- dog

Shuning uchun to‘g‘ri javob faqat bitta class bo‘lmaydi. Bir nechta class birgalikda bo‘lishi mumkin.

```python
# Multi-label classification example labels for one image
image_labels = ["person", "bicycle", "dog"]

print("Labels for this image:", image_labels)
print("Number of labels in this image:", len(image_labels))
```

### Yaxshi Multi-label Dataset

**PASCAL VOC**

- Type: Multi-label classification
- Images bir nechta object classni o‘z ichiga olishi mumkin
- Bitta image ichidagi example labels: `person`, `car`, `dog`
- Nima uchun yaxshi: bir nechta object bor images uchun classic dataset
- Tekshirilgan manba: TensorFlow Datasets VOC page

Links:

- TensorFlow Datasets page: https://www.tensorflow.org/datasets/catalog/voc
- Challenge paper/info page: https://www.microsoft.com/en-us/research/?p=158333
